```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
x=pd.read_csv('data\student_marks (1).csv')
x.drop(["Test_6","Test_7","Test_8","Test_9","Test_10","Test_11","Test_12"],axis=1,inplace=True)
print(x)
print('MODE:')
y=x.mode()
y.drop("Student_ID",axis=1,inplace=True)
y=y.dropna()
print(y)
print('MEAN:')
y=x.mean()
y.drop('Student_ID',axis=0,inplace=True)
print(y)
print('MEDIAN:')
y=x.median()
y.drop('Student_ID',axis=0,inplace=True)
print(y)
x.set_index("Student_ID", inplace=True)
x.plot(kind='line', figsize=(12, 6))
plt.xlabel("Student ID")
plt.ylabel("Test Scores")
plt.title("Test Scores Across Different Tests")
plt.legend(title="Test")
plt.show()
z=x.mean()
z.plot(kind='bar',figsize=(10,6),color=["orange","blue","green","purple","yellow"])
plt.xlabel("Test")
plt.ylabel("Average_Scores")
plt.title("Average Test Scores")
plt.show()
```

```python
z=int(input("Enter number of rounds"))
options=['rock','paper','scissors']
track=[0,0,0]
user=[]
computer=[]
result=[]
for i in range(z):
    x=random.choice(options)#computer
    computer.append(x)
    y=input("Enter choice")#user
    user.append(y)
    if(x=='paper' and y=='scissors') or(x=='rock' and y=='paper') or (x=='scissors' and y=='rock'):
        print("User Won")
        result.append('win')
        track[0]+=1
    elif (x==y):
        print("TIE")
        result.append('tie')
        track[1]+=1
    else:
        print("computer won")
        result.append('loss')
        track[2]+=1
print("User won",track[0],"times","\ncomputer won",track[2],"times","\ntie",track[1],'times')
data={'computer':computer,'user':user,'result':result}
df=pd.DataFrame(data)
df.to_csv('data/game_details.csv')
plt.pie(track,labels=['user','tie','computer'],autopct='%1.1f%%')
plt.title("game records")
plt.legend(loc='lower left')
plt.show()



```
