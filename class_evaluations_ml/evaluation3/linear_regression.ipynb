```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('a.txt')
print(df)
x_data=df['x']
y_data=df['y']
m=np.random.randint(10)
c=np.random.randint(10)
y_pred=m*x_data+c
plt.plot(x_data,y_data,marker='*')
plt.plot(x_data,y_pred,color='red')
```

```python
e=5000
n=len(x_data)
l_r=0.01
for i in range(e):
    y_pred=m*x_data+c
    error=y_pred-y_data
    df_m=(1/n)*np.sum(error*x_data)
    df_c=(1/n)*np.sum(error)
    m=m-l_r*df_m
    c=c-l_r*df_c
print(m,c)
plt.plot(x_data,y_data,marker='*')
plt.plot(x_data,y_pred,color='red')
```
