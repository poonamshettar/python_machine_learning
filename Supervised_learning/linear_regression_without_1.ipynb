```python
import numpy as np
import matplotlib.pyplot as plt
```

2x+1-output

```python
x_data=np.array([x for x in range(20)])
y_data=np.array([2*x+1 for x in range(20)])
print(x_data)
print(y_data)
w1=np.random.randint(10)
b=np.random.randint(10)
y_pred=w1*x_data+b
```

```python
def plot_():
 plt.plot(x_data,y_data,label='real',marker='*')
 plt.plot(x_data,y_pred,linestyle='--',label='predicted')
 plt.legend()
 plt.show()
plot_()
```

```python
m=len(x_data)
def cost_f(x_data,y_data,w1,b):
    y_pred=w1*x_data+b
    error=y_pred-y_data
    k=np.sum(error**2)
    c_f=(1/(2*m))*k
    return c_f
```

```python
e=5000
l_r=0.01
cost=np.zeros(e)
for i in range(e):
    y_pred=w1*x_data+b
    error=y_pred-y_data
    df_w=(1/m)*np.sum(error*x_data)
    df_b=(1/m)*np.sum(error)
    w1=w1-l_r*df_w
    b=b-l_r*df_b
    cost[i]=cost_f(x_data,y_data,w1,b)
plot_()
plt.plot(range(e),cost)
print('weights:',w1)
print('bias:',b)
```
