```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data=np.array([x for x in range(-28,30,2)])
y_data=np.array([2*x**2 for x in range(-28,30,2)])
print(x_data)
print(y_data)
```

```python
w1=np.random.randint(10)
w2=np.random.randint(10)
b=np.random.randint(10)
print(w1,w2,b)
y_pred=w1*x_data+w2*(x_data**2)+b
plt.scatter(x_data,y_data,marker='*',color='red',label='actual')
plt.plot(x_data,y_pred,label='predicted')
plt.legend()
plt.show()
```

```python
e=5000
m=len(x_data)
l_r=0.00001
for i in range(e):
    y_pred=w1*x_data+w2*(x_data**2)+b
    error=y_pred-y_data
    df_w1=(1/m)*(np.sum(error*x_data))
    df_w2=(1/m)*(np.sum(error*(x_data**2)))
    df_b=(1/m)*(np.sum(error))
    w1=w1-l_r*df_w1
    w2=w2-l_r*df_w2
    b-=l_r*df_b
print(w1,w2,b)
```
